```python
wb = Workbook()
ws = wb.active
ws.title = "下载记录"

ws['A1'] = "文件名称"
ws['B1'] = "下载时间（秒）"
ws['C1'] = "文件大小（MB）"
```

```python
# 构建最简单的连接
base_url = "https://asiaeflow.rrd.com:8081/fileDownload/"


download_path = r"D:\work\test\download2"
if not os.path.exists(download_path):
    os.makedirs(download_path)

for row in rows:
    start_time = time.time()
    file_path = row[-1]

    full_path = base_url + file_path

    print(full_path)
    try:
        print("调取请求"+full_path)
        response = requests.get(full_path,verify=False,stream=True)
        if response.status_code == 200:

            filename = os.path.basename(urlparse(full_path).path)
            file_path = os.path.join(download_path,filename)

            with open(file_path, "wb") as f:
                for chunk in response.iter_content(chunk_size=8192):
                    if chunk:
                        f.write(chunk)
                    print(f"下载成功，{filename}")
            end_time = time.time()
            time_row = end_time - start_time

            file_size = os.path.getsize(file_path)/(1024*1024)

            ws.append([
                filename,
                round(time_row,5),
                round(file_size,5)
            ])
            print(f"文件{filename}")
            print(f"时间{time_row}")
            print(f"文件大小{file_size}")


        else:
            print(f"下载失败；状态码：{response.status_code}")
    except:
        print("代码错误")
```
